```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
```

```python
Scaled_cc = cc.apply(lambda x:(x.astype(float) - min(x))/(max(x)-min(x)), axis = 0)
```

```python
Scaled_cc
```

```python
from sklearn.cluster import KMeans
plt.rcParams["figure.figsize"] = 10,8
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 500, n_init = 20, random_state = 0)
    kmeans.fit(Scaled_cc)
    wcss.append(kmeans.inertia_)
    
#Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('Optimum K for CC General')
plt.xlabel('Agrupaciones')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
wcss
```

```python
r =[]
for v in range(len(wcss)):
    if v+1 < len(wcss):
        r.append((wcss[v]-wcss[v+1])/(wcss[v]))
r
```
